# 2. Methodology

## 2.1 KMeans

The methodology used for data analysis is the Machine Learning technique called KMeans.

Based on this technique and based on the data presented, it segments a group of data based on the average distance from each point to another point.
Choosing and aggregating the groups of points that have greater similarity and separating the data that have greater dissimilarity.
The distance is calculated by averaging the features of each data, which improves the grouping process by similarity.

For this capstone, the data were grouped into 3 clusters, with some similarities.

To achieve the objective of this analysis, which is to choose the best neighborhood to live in as a person who likes coffee, the algorithm segmented more than 240 samples collected via the Foursquare API and grouped them based on similarities between them, thus generating 3 clusters.
